# Clean the data

```python
df_train.loc[:, "missing_info"] = df_train["missing_info"].astype(str)
df_train.loc[:, "improvements"] = df_train["improvements"].astype(str)
df_train.loc[:, "missing_info_en"] = df_train["missing_info_en"].astype(str)
df_train.loc[:, "improvements_en"] = df_train["improvements_en"].astype(str)
```

```python
import re
import string

#--------------------------------------------------
punctuations = list(string.punctuation)
def listToDict(lst):
    return {lst[i]: '' for i in range(0, len(lst), 2)}

punctuations =  listToDict(punctuations)
#--------------------------------------------------

def clean(string):
    if string in ["nan", "#VALUE!"]:
        return ""
    string = string.lower()
    
    #Remove tildes
    vocales = {"á": "a", "é": "e", "í": "i", "ó": "o", "ú": "u"}
    for vocal in vocales.keys():
        string = string.replace(vocal, vocales[vocal])
        
    #Remove e-mails: user@xxx.com
    email = '\S*@\S*\s?'
    pattern = re.compile(email)
    pattern.sub('', string)
    
    #Remove urls
    re.sub(r'http\S+','',string)
    
    #Remove non-alphanumeric characters
    for punctuation in punctuations.keys():
        string = string.replace(punctuation, punctuations[punctuation])
    
    #Remove punctuation
    string = re.sub(r'[^(a-zA-Z)\s]','', string)
    return string
```

```python
df_train.loc[:, 'missing_info'] =  df_train['missing_info'].apply(lambda x: clean(x))
df_train.loc[:, 'improvements'] =  df_train['improvements'].apply(lambda x: clean(x))
df_train.loc[:, 'missing_info_en'] =  df_train['missing_info_en'].apply(lambda x: clean(x))
df_train.loc[:, 'improvements_en'] =  df_train['improvements_en'].apply(lambda x: clean(x))
```

# Tokenizing and removing stopwords

```python
import nltk
nltk.download('stopwords')
#nltk.download('punkt')
```

```python
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def remove_stop_words(string, language):
    if string == "":
        return ""
    # Create dictionary of stopwords for spanish and english
    stopwords_dict = {
        "es": list(set(stopwords.words('spanish'))),
        "en": list(set(stopwords.words('english')))
    }
    # tokenize 
    tokenized = word_tokenize(string)
    # remove stopwords 
    stopped = [w for w in tokenized if not w in stopwords_dict[language]]
    # join the list of above words to create a sentence without stop words
    filtered_string = (" ").join(stopped)
    return filtered_string
```

```python
df_train.loc[:, 'missing_info_wo_stop_words'] =  df['missing_info'].apply(lambda x: remove_stop_words(x, 'es'))
df_train.loc[:, 'improvements_wo_stop_words'] =  df['improvements'].apply(lambda x: remove_stop_words(x, 'es'))
df_train.loc[:, 'missing_info_en_wo_stop_words'] =  df['missing_info_en'].apply(lambda x: remove_stop_words(x, 'en'))
df_train.loc[:, 'improvements_en_wo_stop_words'] =  df['improvements_en'].apply(lambda x: remove_stop_words(x, 'en'))
```

# Create Bag of Words (BOW)

```python
# dictionary to hold english pos tags and their explanations
pos_dict = {
'CC': 'coordinating conjunction',
'CD': 'cardinal digit',
'DT': 'determiner',
'EX': 'existential there (like: \"there is\" ... think of it like \"there exists\")',
'FW': 'foreign word',
'IN':  'preposition/subordinating conjunction',
'JJ': 'adjective \'big\'',
'JJR': 'adjective, comparative \'bigger\'',
'JJS': 'adjective, superlative \'biggest\'',
'LS': 'list marker 1)',
'MD': 'modal could, will',
'NN': 'noun, singular \'desk\'',
'NNS': 'noun plural \'desks\'',
'NNP': 'proper noun, singular \'Harrison\'',
'NNPS': 'proper noun, plural \'Americans\'',
'PDT': 'predeterminer \'all the kids\'',
'POS': 'possessive ending parent\'s',
'PRP': 'personal pronoun I, he, she',
'PRP$': 'possessive pronoun my, his, hers',
'RB': 'adverb very, silently,',
'RBR': 'adverb, comparative better',
'RBS': 'adverb, superlative best',
'RP': 'particle give up',
'TO': 'to go \'to\' the store.',
'UH': 'interjection errrrrrrrm',
'VB': 'verb, base form take',
'VBD': 'verb, past tense took',
'VBG': 'verb, gerund/present participle taking',
'VBN': 'verb, past participle taken',
'VBP': 'verb, sing. present, non-3d take',
'VBZ': 'verb, 3rd person sing. present takes',
'WDT': 'wh-determiner which',
'WP': 'wh-pronoun who, what',
'WP$': 'possessive wh-pronoun whose',
'WRB': 'wh-abverb where, when',
'QF' : 'quantifier, bahut, thoda, kam (Hindi)',
'VM' : 'main verb',
'PSP' : 'postposition, common in indian langs',
'DEM' : 'demonstrative, common in indian langs'
}

def extract_pos(doc):
    parsed_text = {'word':[], 'pos':[], 'exp':[]}
    for sent in doc.sentences:
        for wrd in sent.words:
            if wrd.pos in pos_dict.keys():
                pos_exp = pos_dict[wrd.pos]
            else:
                pos_exp = 'NA'
            parsed_text['word'].append(wrd.text)
            parsed_text['pos'].append(wrd.pos)
            parsed_text['exp'].append(pos_exp)
    return pd.DataFrame(parsed_text)
```

```python
# Function to extract either only adjectives or noun plus adjectives
def extract_adj_noun(txt, language="en", noun=False):
    if language == "es":
        corpus = nlp_es(txt)
    else:
        corpus = nlp_en(txt)
    BOW = extract_pos(corpus)
    adj_pos = ["ADJ", "JJ", "JJR", "JJS"]
    noun_pos = ["NOUN", "NN", "NNS", "NNP", "NNPS"]
    if noun:
        return " ".join(BOW[BOW["pos"].isin(adj_pos + noun_pos)]["word"])
    return " ".join(BOW[BOW["pos"].isin(adj_pos)]["word"])
```

```python
# Get spanish adjectives and nouns + adjectives
df_train.loc[:, "missing_info_adj"] = df_train["missing_info_wo_stop_words"].apply(lambda x: extract_adj_noun(x, 'es') if x != "" else "")
df_train.loc[:, "improvements_adj"] = df_train["improvements_wo_stop_words"].apply(lambda x: extract_adj_noun(x, 'es') if x != "" else "")
df_train.loc[:, "missing_info_noun_adj"] = df_train["missing_info_wo_stop_words"].apply(lambda x: extract_adj_noun(x, 'es', True) if x != "" else "")
df_train.loc[:, "improvements_noun_adj"] = df_train["improvements_wo_stop_words"].apply(lambda x: extract_adj_noun(x, 'es', True) if x != "" else "")
```

```python
# Get english adjectives and nouns + adjectives
df_train.loc[:, "missing_info_en_adj"] = df_train["missing_info_en_wo_stop_words"].apply(lambda x: extract_adj_noun(x, 'en') if x != "" else "")
df_train.loc[:, "improvements_en_adj"] = df_train["improvements_en_wo_stop_words"].apply(lambda x: extract_adj_noun(x, 'en') if x != "" else "")
df_train.loc[:, "missing_info_en_noun_adj"] = df_train["missing_info_en_wo_stop_words"].apply(lambda x: extract_adj_noun(x, 'en', True) if x != "" else "")
df_train.loc[:, "improvements_en_noun_adj"] = df_train["improvements_en_wo_stop_words"].apply(lambda x: extract_adj_noun(x, 'en', True) if x != "" else "")
```
