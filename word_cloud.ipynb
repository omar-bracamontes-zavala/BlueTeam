```python
import pandas as pd

adj_en = pd.read_csv("datasets/adjs_en.csv").fillna('')
adj_es = pd.read_csv("datasets/adjs_es.csv").fillna('')
df = pd.read_csv("datasets/dataset_oficial_adj_nouns.csv")
```

```python
adj_en = adj_en.loc[adj_en["word"].str.len() > 2]
adj_es = adj_es.loc[adj_es["word"].str.len() > 2]
```

```python
text_en = " ".join(adj_en["word"])
text_es = " ".join(adj_es["word"])
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

wordcloud_en = WordCloud(max_words=30, background_color="white").generate(text_en)
wordcloud_es = WordCloud(max_words=30, background_color="white").generate(text_es)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(wordcloud_en, interpolation="bilinear")
plt.title('Texto en inglés')
plt.axis("off")
plt.show()
```

```python
plt.figure()
plt.imshow(wordcloud_es, interpolation="bilinear")
plt.title('Texto en español')
plt.axis("off")
plt.show()
```

```python
institutions = df["institution"].str.split('-', expand=True)
```
